```python
### Test 1: Random Data

# Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# Labels for x-axis
time = [100,1000,10000,100000,1000000]

# Bar values for different sorting algorithms
red = [71,449,13328,0,0] # Results that are difficult to plott [71,449,13328,4130000,0]
orange = [33,88,608,9135,113583]
yellow = [68,373,1538,16752,185304]
green = [52,132,1047,10504,126913]
blue = [73,470,1547,16621,192632]
purple = [17,442,885,14437,75180]

# Define positions of bars on x-axis
x_positions = np.arange(len(time))

# Set the figure size
fig, ax = plt.subplots(figsize=(10,6))

# Plot the bars with different colors and alpha values
ax.bar(x_positions, red, color='red', alpha=0.8, width=0.1)
ax.bar(x_positions + 0.1, orange, color='orange', alpha=0.8, width=0.1)
ax.bar(x_positions + 0.2, yellow, color='yellow', alpha=0.8, width=0.1)
ax.bar(x_positions + 0.3, green, color='green', alpha=0.8, width=0.1)
ax.bar(x_positions + 0.4, blue, color='blue', alpha=0.8, width=0.1)
ax.bar(x_positions + 0.5, purple, color='purple', alpha=0.8, width=0.1)

# Set the x-axis ticks and labels, and label the x- and y-axis
ax.set_xticks(x_positions + 0.25)
ax.set_xticklabels(time)
ax.set_xlabel('Problem size [Number of elements]')
ax.set_ylabel('Running Time [µs]')

# Set the title of the plot and adjust the y-axis limits
ax.set_title('Test 1: Random Data')
ax.set_ylim([0, 200000])

# Add a legend to the plot
ax.legend(['InsertionSort', 'Qs V1', 'Qs V2', 'Qs V3', 'Qs V4', 'Arrays.sort'])

# Display the plot
plt.show()
```
